```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nltk import word_tokenize
import nltk
```

```python
train = pd.read_json('train.json')
```

```python
train.head()
```

```python
bag_of_words = []
for recipe in train.ingredients:
    bag_of_words += recipe

bag_of_words = nltk.FreqDist(bag_of_words)
bag_of_words.most_common()[-20:-1]
```

```python
fig, axes = plt.subplots(nrows=1, sharey=False, figsize=(10,5))
bag_of_words.plot(15, cumulative = False)
```

```python
fig, axes = plt.subplots(nrows=1, sharey=False, figsize=(10,5))
g = sns.countplot(train.cuisine, color = 'green')
g.set_xlabel("Cuisine",fontsize=12)
g.set_ylabel("Total",fontsize=12)
g.set_title("Distribution of Recipes by Cuisine",fontsize=20)
g.tick_params(labelsize=15)
g.set_xticklabels(g.get_xticklabels(),rotation=45,ha="right",rotation_mode='anchor')
```

```python
train['count_ingredients'] = train.ingredients.apply(len)
```

```python
fig, axes = plt.subplots(nrows=1, sharey=False, figsize=(10,5))
g = sns.distplot(train[train.count_ingredients <= 30].count_ingredients)
g.set_xlabel("Total Ingredients per Recipe",fontsize=12)
g.set_title("Distribution of Ingredients per Recipe",fontsize=20)
```
